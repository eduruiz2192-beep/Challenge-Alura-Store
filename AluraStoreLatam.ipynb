### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
facturacion_t1 = tienda['Precio'].sum()
facturacion_t2 = tienda2['Precio'].sum()
facturacion_t3 = tienda3['Precio'].sum()
facturacion_t4 = tienda4['Precio'].sum()

print(f"Facturación Tienda 1: ${facturacion_t1:,.2f}")
print(f"Facturación Tienda 2: ${facturacion_t2:,.2f}")
print(f"Facturación Tienda 3: ${facturacion_t3:,.2f}")
print(f"Facturación Tienda 4: ${facturacion_t4:,.2f}")
```

# 2. Ventas por categoría

```python
print("--- Categorías Tienda 1 ---")
print(tienda['Categoría del Producto'].value_counts())
print("\n--- Categorías Tienda 2 ---")
print(tienda2['Categoría del Producto'].value_counts())
print("\n--- Categorías Tienda 3 ---")
print(tienda3['Categoría del Producto'].value_counts())
print("\n--- Categorías Tienda 4 ---")
print(tienda4['Categoría del Producto'].value_counts())
```

# 3. Calificación promedio de la tienda


```python
calif_t1 = tienda['Calificación'].mean()
calif_t2 = tienda2['Calificación'].mean()
calif_t3 = tienda3['Calificación'].mean()
calif_t4 = tienda4['Calificación'].mean()

print(f"Calificación Promedio Tienda 1: {calif_t1:.2f} estrellas")
print(f"Calificación Promedio Tienda 2: {calif_t2:.2f} estrellas")
print(f"Calificación Promedio Tienda 3: {calif_t3:.2f} estrellas")
print(f"Calificación Promedio Tienda 4: {calif_t4:.2f} estrellas")
```

# 4. Productos más y menos vendidos

```python
print("--- Top 3 Más Vendidos ---")
print("Tienda 1:\n", tienda['Producto'].value_counts().head(3))
print("\nTienda 2:\n", tienda2['Producto'].value_counts().head(3))
print("\nTienda 3:\n", tienda3['Producto'].value_counts().head(3))
print("\nTienda 4:\n", tienda4['Producto'].value_counts().head(3))

print("\n" + "="*30 + "\n")

print("--- Top 3 Menos Vendidos ---")
print("Tienda 1:\n", tienda['Producto'].value_counts().tail(3))
print("\nTienda 2:\n", tienda2['Producto'].value_counts().tail(3))
print("\nTienda 3:\n", tienda3['Producto'].value_counts().tail(3))
print("\nTienda 4:\n", tienda4['Producto'].value_counts().tail(3))
```

# 5. Envío promedio por tienda

```python
envio_t1 = tienda['Costo de envío'].mean()
envio_t2 = tienda2['Costo de envío'].mean()
envio_t3 = tienda3['Costo de envío'].mean()
envio_t4 = tienda4['Costo de envío'].mean()

print(f"Costo de Envío Promedio Tienda 1: ${envio_t1:,.2f}")
print(f"Costo de Envío Promedio Tienda 2: ${envio_t2:,.2f}")
print(f"Costo de Envío Promedio Tienda 3: ${envio_t3:,.2f}")
print(f"Costo de Envío Promedio Tienda 4: ${envio_t4:,.2f}")
```

## 6. Visualización de Datos (Generando gráficos)

```python
import matplotlib.pyplot as plt

# --- Datos que ya calculamos ---
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
facturacion = [facturacion_t1, facturacion_t2, facturacion_t3, facturacion_t4]
envios = [envio_t1, envio_t2, envio_t3, envio_t4]
calificaciones = [calif_t1, calif_t2, calif_t3, calif_t4]

# --- Gráfico 1: Barras (Facturación Total) ---
plt.figure(figsize=(8, 4))
plt.bar(tiendas, facturacion, color=['#4C72B0', '#55A868', '#C44E52', '#8172B3'])
plt.title('1. Facturación Total por Tienda', fontsize=14)
plt.ylabel('Ingresos Totales ($)')
plt.show()

# --- Gráfico 2: Líneas (Costo de Envío Promedio) ---
plt.figure(figsize=(8, 4))
plt.plot(tiendas, envios, marker='o', linestyle='-', color='#D55E00', linewidth=2, markersize=8)
plt.title('2. Costo de Envío Promedio', fontsize=14)
plt.ylabel('Costo de Envío ($)')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

# --- Gráfico 3: Dispersión (Calificación Promedio) ---
plt.figure(figsize=(8, 4))
plt.scatter(tiendas, calificaciones, color='#009E73', s=150) # s es el tamaño del punto
plt.title('3. Calificación Promedio de Clientes', fontsize=14)
plt.ylabel('Estrellas (1 a 5)')
plt.ylim(3.5, 4.5) # Ajustamos el eje Y para ver mejor la diferencia
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```
